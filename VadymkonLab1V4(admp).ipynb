```python
import random
def sort_list(n) : 
  list1=[ random.random() for _ in range(n)]
  list1.sort()
  return list1

print(sort_list(50)) 
```

```python
import numpy as np
import random

def sort_np (n): 
  arr1 = np.random.rand(n)
  arr1.sort()
  return arr1
arr = sort_np(50)
arr
```

```python
%%timeit 
list1 = sort_list(5000)
```

```python
%%timeit -r 15 -n 15
arr = sort_np(5000)
```

```python
import random
def my_sort_list(n):
  arr = [ random.random() for _ in range(n)]
  for i in range(0, len(arr)):
      for j in range(0, len(arr) - i - 1):
          if arr[j] > arr[j + 1]:
              buff = arr[j]
              arr[j] = arr[j + 1]
              arr[j + 1] = buff
  return arr
%timeit -r 15 -n 15 my_sort_list(500)
```

```python
import numpy as np
import random

def my_sort_np (n): 
  arr = np.random.rand(n)
  for i in range(0, len(arr)):
      for j in range(0, len(arr) - i - 1):
          if arr[j] > arr[j + 1]:
              buff = arr[j]
              arr[j] = arr[j + 1]
              arr[j + 1] = buff
  return arr
my_sort_np(10)
```

```python
%%timeit 
list1 = my_sort_list(50)
```

```python
%%timeit -r 15 -n 15
arr = my_sort_np(50)
```

```python
t = %timeit -r 10 -o arr = my_sort_np(50)
print(t.all_runs)
print(min(t.all_runs))
t = %timeit -r 10 -o arr = my_sort_list(50)
print(t.all_runs)
print(min(t.all_runs))
t = %timeit -r 10 -o arr = sort_np(50)
print(t.all_runs)
print(min(t.all_runs))
```
