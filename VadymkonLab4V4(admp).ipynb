```python
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.decomposition import PCA
```

```python
p=pd.read_csv('/content/points.csv')
p
```

```python
p.info()
```

```python
p.describe()
```

```python
p.var()
```

```python
p.corr()
```

```python
fig = px.scatter_3d(p, x='x', y='y', z='z')
fig.show()
```

```python
fig.update_traces(marker={'size':3})
fig.update_layout(scene={'aspectmode': 'data'}) 
#масштаб
```

```python
pca = PCA()
pca.fit(p)
```

```python
p_trans=pd.DataFrame(pca.transform(p), columns = ["c1", "c2", "c3"])
p_trans
```

```python
p_trans.describe()
```

```python
p_trans.var()
```

```python
p.var()
```

```python
p_trans.corr() #ближче до нуля, що були не 1
```

```python
p.corr()
```

```python
fig_trans = px.scatter_3d(p_trans, x = "c1", y = "c2", z = "c3")
fig_trans.update_traces(marker={"size":3})
fig_trans.update_layout(scene = {"aspectmode": "data"})
```

```python
p_trans_dimred  = p_trans.copy()
p_trans_dimred['c3'] = 0
p_trans_dimred
```

```python
fig_trans.add_scatter3d(x=p_trans_dimred.c1,
                        y=p_trans_dimred.c2,
                        z=p_trans_dimred.c3,
                        mode='markers', marker_size=3)
```

```python
p_dimred=pd.DataFrame(pca.inverse_transform(p_trans_dimred), columns = ["x", "y", "z"])
p_dimred
```

```python
fig.add_scatter3d(      x=p_dimred.x,
                        y=p_dimred.y,
                        z=p_dimred.z,
                        mode='markers', marker_size=3)
```
